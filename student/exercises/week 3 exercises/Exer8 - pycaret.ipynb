## Palmer Penguins Dataset

The goal of palmer penguins is to provide a great dataset for data exploration & visualization, as an alternative to iris.

The data contains 344 penguins. There are 3 different species of penguins in this dataset, collected from 3 islands in the Palmer Archipelago, Antarctica.

- Features: `4` numerical, `2` categorical
- Target: `species` (Categorical / 3 classes)
- Size: `344` samples
- Source: [Palmer Penguins](https://allisonhorst.github.io/palmerpenguins/)

### Load the data

```python
import pandas as pd

try:
    import seaborn as sns
    penguins = sns.load_dataset('penguins')
except ModuleNotFoundError:
    penguins = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')

penguins
```

## Experiments

1. **First experiment**:
    - X = Flipper Length (numerical) & Bill Length (numerical)
    - y = Species (categorical)
2. **Second experiment**:
    - X = Weights (numerical) & Species (categorical)
    - y = Sex (categroical)
3. **Third experiment**:
    - X = `island`, `bill_length_mm`, `bill_depth_mm`, `flipper_length_mm`, `body_mass_g`
    - y = Sex and Species (PyCaret needs one target column, so we will combine them into one label)

```python
from IPython.display import display
from pycaret.classification import compare_models, plot_model, predict_model, pull, save_model, setup

# First experiment: predict penguin species from flipper length and bill length.
exp1_data = penguins[["flipper_length_mm", "bill_length_mm", "species"]].dropna().copy()

exp1 = setup(
    data=exp1_data,
    target="species",
    session_id=42,
    n_jobs=1,
    html=False,
    verbose=False,
)

best_species_model = compare_models()
exp1_leaderboard = pull()
display(exp1_leaderboard.head())

plot_model(best_species_model, plot="confusion_matrix")

exp1_predictions = predict_model(best_species_model, data=exp1_data.head())
display(exp1_predictions)

save_model(best_species_model, "penguins_species_model")
```

```python
# Second experiment: predict penguin sex from body mass and species.
exp2_data = penguins[["body_mass_g", "species", "sex"]].dropna().copy()

exp2 = setup(
    data=exp2_data,
    target="sex",
    session_id=42,
    n_jobs=1,
    html=False,
    verbose=False,
)

best_sex_model = compare_models()
exp2_leaderboard = pull()
display(exp2_leaderboard.head())

plot_model(best_sex_model, plot="confusion_matrix")

exp2_predictions = predict_model(best_sex_model, data=exp2_data.head())
display(exp2_predictions)

save_model(best_sex_model, "penguins_sex_model")
```

```python
# Third experiment: combine species and sex into one multiclass target.
exp3_data = penguins[[
    "island",
    "bill_length_mm",
    "bill_depth_mm",
    "flipper_length_mm",
    "body_mass_g",
    "species",
    "sex",
]].dropna().copy()

exp3_data["species_sex"] = exp3_data["species"] + " | " + exp3_data["sex"]
exp3_data = exp3_data.drop(columns=["species", "sex"])

exp3 = setup(
    data=exp3_data,
    target="species_sex",
    session_id=42,
    n_jobs=1,
    html=False,
    verbose=False,
)

best_species_sex_model = compare_models()
exp3_leaderboard = pull()
display(exp3_leaderboard.head())

plot_model(best_species_sex_model, plot="confusion_matrix")

exp3_predictions = predict_model(best_species_sex_model, data=exp3_data.head())
display(exp3_predictions)

save_model(best_species_sex_model, "penguins_species_sex_model")
```
