```python
import seaborn as sns
import pandas as pd
from scipy.stats import chi2_contingency, spearmanr, mannwhitneyu
import statsmodels.formula.api as smf

tips = sns.load_dataset("tips")
alpha = 0.05
```

### Hypothesis H1: smoking is associated with time of visit


```python
# so H1: smoking is associated with time of visit and H0: smoker and time are independent
ct = pd.crosstab(tips["smoker"], tips["time"])
chi2, p1, dof, expected = chi2_contingency(ct)

print("H1 contingency table:")
print(ct)
print(f"H1 p-value = {p1:.4f}")
```

```python
# since p-value = 0.4771 > alpha = 0.05, we fail to reject the null hypothesis
# there is no significant association between smoking and time of visit
```

### Hypothesis H2: the bigger the group the higher the tip


```python
# so H2: the bigger the group the higher the tip
# and H0: no positive association between size and tip

rho, p2 = spearmanr(tips["size"], tips["tip"], alternative="greater")

print(f"\nH2 Spearman rho = {rho:.4f}")
print(f"H2 p-value = {p2:.4g}")
```

```python
# since p-value = 5.299e-15 <alpha = 0.05, we fail to reject the alternative hypothesis
# there is a significant association between size and tip the bigger the group the higher the tip
```

### Hypothesis H3: group size is different based on the time of visit


```python
# so, H3: group size is different based on time of visit
# and H0: size has the same distribution at Lunch and Dinner
lunch = tips.loc[tips["time"] == "Lunch", "size"]
dinner = tips.loc[tips["time"] == "Dinner", "size"]
u_stat, p3 = mannwhitneyu(lunch, dinner, alternative="two-sided")
print(f"\nLunch mean size = {lunch.mean():.3f}")
print(f"Dinner mean size = {dinner.mean():.3f}")
print(f"H3 p-value = {p3:.4f}")
```

```python
# since the p-value = 0.0102 < alpha = 0.05, we fail to reject the alternative hypothesis
# there is a significant difference in group size based on time of visit, with dinner groups being
```

### Hypothesis H4: tip percentage differs between smokers and non-smokers


```python
# so, H4: tip_pct dont have the same distribution for smokers and non-smokers
# and H0: tip_pct has the same distribution for smokers and non-smokers
tips["tip_pct"] = tips["tip"] / tips["total_bill"]

smoke_yes = tips.loc[tips["smoker"] == "Yes", "tip_pct"]
smoke_no = tips.loc[tips["smoker"] == "No", "tip_pct"]
u_stat4, p4 = mannwhitneyu(smoke_yes, smoke_no, alternative="two-sided")

print(f"\nSmoker mean tip % = {smoke_yes.mean():.4f}")
print(f"Non-smoker mean tip % = {smoke_no.mean():.4f}")
print(f"H4 p-value = {p4:.4f}")
```

```python
# since the p-value = 0.5607 > alpha = 0.05, we fail to reject the null hypothesis
# there is no significant difference in tip percentage between smokers and non-smokers
```

### Confounder / mediation check for size

```python

# Model 1: tip explained by size only
m1 = smf.ols("tip ~ size", data=tips).fit()

# Model 2: tip explained by size and total_bill
m2 = smf.ols("tip ~ size + total_bill", data=tips).fit()

# Model 3: total_bill explained by size
m3 = smf.ols("total_bill ~ size", data=tips).fit()

print("\nModel 1 coefficients:")
print(m1.params)
print(m1.pvalues)

print("\nModel 2 coefficients:")
print(m2.params)
print(m2.pvalues)

print("\nModel 3 coefficients:")
print(m3.params)
print(m3.pvalues)
```
